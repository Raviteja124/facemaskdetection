```python
#cv2.rectangle(img,(x,y),(w,h),(b,g,r),border_thickness)
import cv2
img=cv2.imread(r"C:\Users\teja1\Downloads\priya1.jpeg")
haar=cv2.CascadeClassifier('data.xml')
while True:
    faces=haar.detectMultiScale(img)
    for x,y,w,h in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,255),5)
    cv2.imshow('result',img)
    if cv2.waitKey(2)==27:
        break
cv2.destroyAllWindows()
```

```python
cap=cv2.VideoCapture(0)
data=[]
while True:
    flag,img=cap.read()
    if flag:
        faces=haar.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,255),4)
            face=img[y:y+h,x:x+w,:]
            face=cv2.resize(face,(50,50))
            print(len(data))
            if(len(data)<400):
                data.append(face)
        cv2.imshow('result',img)
        if cv2.waitKey(2)==27 or len(data)>=200:
            break
cap.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
np.save('withoutmask.npy',data)
```
